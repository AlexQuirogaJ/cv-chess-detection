# Get dataset images

```python
def video_from_folder_images(folder_path, video_name, fps=2):

    snapshots = []
    for img_path in sorted(pathlib.Path(folder_path).iterdir()):
        if img_path.suffix.lower() not in [".jpg", ".jpeg", ".png"]:
            continue

        iter_number = int(img_path.stem.split("_")[-1])
        img = cv2.imread(str(img_path))

        # Add frame number as text
        show_text = f"Move: {iter_number}"

        cv2.putText(
            img,
            show_text,
            (20, 40),  # Position of the text
            cv2.FONT_HERSHEY_SIMPLEX,
            0.7,
            (255, 255, 255),
            2,
        )

        snapshots.append(img)

    output_video_path = f"{folder_path}/{video_name}.mp4"
    save_video(snapshots, output_video_path, fps)
    
def save_video(output_video_frames, output_video_path, fps=2):
    output_dir = output_video_path.rsplit("/", 1)[0]
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    fourcc = cv2.VideoWriter_fourcc(
        *"MJPG"
    )  # Sets the video codec using OpenCV's VideoWriter_fourcc
    # MJPG is a codec (Motion JPEG) and * unpacks the string so it's passed as separate characters to the function
    out = cv2.VideoWriter(
        output_video_path,
        fourcc,  #
        fps,  # 24 FPS
        (
            output_video_frames[0].shape[1],
            output_video_frames[0].shape[0],
        ),  # (width, height)
    )
    for frame in output_video_frames:
        out.write(frame)
    out.release()
```

```python
current_dir = pathlib.Path().cwd()
root_dir = current_dir.parent

recorded_dataset_dir = root_dir / "data" / "recorded_dataset"
recorded_dataset_dir.mkdir(parents=True, exist_ok=True)

# video_id = "rrPfmSWlAPM"
# time_stamps = time_stamps_rrPfmSWlAPM

video_id = "JSnfTfK__Bc"
time_stamps = time_stamps_JSnfTfK__Bc

video_path = recorded_dataset_dir / video_id / f"{video_id}.mov"
recorded_dataset_dir_images = recorded_dataset_dir / video_id / "irl"
recorded_dataset_dir_images.mkdir(parents=True, exist_ok=True)

vidcap = cv2.VideoCapture(video_path)
frame_count = vidcap.get(cv2.CAP_PROP_FRAME_COUNT)
fps = vidcap.get(cv2.CAP_PROP_FPS)

for idx, time_stamp in enumerate(time_stamps):
    time_stamp = sum(
        x * int(t) for x, t in zip([60, 1, 0.01], time_stamp.split(":"))
    )
    vidcap.set(cv2.CAP_PROP_POS_MSEC, time_stamp * 1000)

    success, image = vidcap.read()
    if success:
        # Rotate image 90 degrees clockwise
        image = cv2.rotate(image, cv2.ROTATE_90_CLOCKWISE)

        # Save frame as JPEG file
        cv2.imwrite(
            str(recorded_dataset_dir_images / f"{video_id}_{idx:04d}.jpg"),
            image,
                )
    
# Save video from images
video_from_folder_images(
    folder_path=str(recorded_dataset_dir_images),
    video_name=f"{video_id}_from_images",
    fps=1,
)
```
